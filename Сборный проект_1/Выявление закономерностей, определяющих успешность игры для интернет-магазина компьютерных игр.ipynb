# Выявление закономерностей, определяющих успешность игры для интернет-магазина компьютерных игр.


В нашем распоржении из открытых источников исторические данные до 2016 года о продажах игр, оценки пользователей и экспертов, жанры и платформы. Необхоимо выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании интернет-магазина «Стримчик», который продаёт по всему миру компьютерные игры.


**Цель исследования** 
- определить для пользователя каждого региона (NA, EU, JP):
1. самые популярные платформы (топ-5);
2. самые популярные жанры (топ-5);
3. влияет ли рейтинг ESRB на продажи в отдельном регионе.
- проверить 2 гипотезы:
1. средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
2. средние пользовательские рейтинги жанров Action (англ. «действие», экшенигры) и Sports (англ. «спортивные соревнования») разные.


**Ход исследования**

Исторические данные получим из файла `/datasets/games.csv`

О качестве данных ничего не известно. Поэтому перед подготовкой ответов понадобится обзор данных.
Проверим данные на ошибки и оценим их влияние на исследование. Затем, на этапе предобработки исправим самые критичные ошибки данных.
 
Таким образом, исследование пройдёт в пять этапов:
   1. Обзор данных.
   2. Подготовка данных.
   3. Исследовательский анализ данных
   4. Составление портрета пользователя каждого региона
   5. Проверка гипотез.

### Обработка пропусков

Сначала посчитаем, сколько в таблице пропущенных значений. Для этого достаточно двух методов `pandas`:

```python
display(data.isna().sum()) # подсчёт пропусков
```

При рассмотрении пропусков были выявлены две игры без названия, жанра и рейтингов. Данные были внесены не в полном объеме в связи с тем, что по платформе GEN в 1993 году не вносилось название и жанр игр при продаже менее 50 тыс. копий или продаже за пределами Японии, а рейтингование появилось позже. Данный объем данных не оказывает значительного влияние на наше исследование и составляет 0.01%. Поэтому с легкостью удалим их.

```python
display(data[data['name'].isna()]) # отражение пропусков по столбцу 'name'
```

```python
display(data[(data['year_of_release'] == 1993.0) & (data['platform'] == 'GEN')]) # срез по платформе GEN в 1993 году
```

```python
data = data.dropna(subset=['name']).reset_index(drop=True) 
# удаление строк с пустыми значениями в столбце 'name' (с удалением старых индексов и формированием новых)
```

Перейдем к пропускам в столбце `year_of_release`. Они могли возникнуть в связи с человеческим фактором при занесении и обработке данных. 

Для начала создадим новую переменную `transformed_year`, в которой сгруппируем данные по названию игр и подсчитаем средний год выпуска каждой группы, предварительно его огруглив до ближайшего целого значения.

```python
transformed_year = round(data.groupby('name')['year_of_release'].transform('mean'))
# вычисление среднего года выпуска каждой игры с округлением до ближайшего целого
```

```python
data['year_of_release'] = data['year_of_release'].fillna(transformed_year) 
# замена пропусков в столбце `year_of_release` на средний год выпуска каждой игры
```

```python
print(data['year_of_release'].isna().sum()) # подсчёт пропусков в столбце 'year_of_release'
```

После проведения замены в столбце `year_of_release` ещё осталось 146 пропусков. Данный объем данных не оказывает значительного влияние на наше исследование и составляет 0.9%. Поэтому без мучений совести удалим строки, с оставшимися пропусками.

```python
data = data.dropna(subset=['year_of_release']).reset_index(drop=True) 
# удаление строк с пустыми значениями в столбце 'year_of_release'(с удалением старых индексов и формированием новых)
```

Рейтинг от организации ESRB был основан только в 1994 году, только для США и Канады и данные наполнялись постепенно. Также рейтингование компьютерных игр критиками и пользователями начало набирать обороты только в начале XXI века с ростом их доступности и популярности. Соответственно, это объясняет большое количество пропусков в столбцах с рейтингами. 

Для того чтобы не потерять большой объем данных и удобства анализа необходимо принять решение что делать с пропусками в 3 столбцах: `critic_score`, `user_score` и `rating`.

В столбце `critic_score` заменим пропущенные значения двумя этапами:
1. Создадим новую переменную `transformed_critic_score`, в которой сгруппируем данные по названию игр и подсчитаем среднюю оценку критиков для каждой группы, предварительно её огруглив до ближайшего целого значения. Произведем замену пропусков значением из этой переменной при совпадении названия игры.
2. Для замены оставшихся 7532 пропусков используем нулевое значение. При выгрузке уникальных значений был выявлен диапазон оценок от 13 до 98. Соответственно, -1 даст возможность учитывать при анализе полный спектр строк, сразу определить строки замены и минимизировать искажение. 

```python
transformed_critic_score = round(data.groupby('name')['critic_score'].transform('mean'))
# вычисление средней оценки для названия каждой игры с округлением до ближайшего целого
data['critic_score'] = data['critic_score'].fillna(transformed_critic_score) 
# замена пропусков в столбце 'critic_score' на среднюю оценку для названия каждой игры
```

```python
print(data['critic_score'].isna().sum()) # подсчёт пропусков после первой замены в столбце 'critic_score'
```

```python
print(data['critic_score'].sort_values().unique()) # выгрузка уникальных значений оценок критиков с сортировкой по возрастанию
```

```python
data['critic_score'] = data['critic_score'].fillna(-1) # заполнение пропусков нулевым значением в столбце 'critic_score'
```

В столбце `user_score` (помимо пропусков) было обнаружено 2394 значений `tbd`. Чаще всего данная аббревиатура используется для обозначения неизвестной даты выхода игры в продажу. Но может быть использована и в случае с уже вышедшей игрой, если у них готовится к релизу какое-либо дополнение, однако пока не известно, когда именно оно будет выпущено. Но в данном случае данная аббревиатура применяется для тех элементов, которые не имеют по крайней мере четырех (4) отзывов в базе данных. Соответственно, в нашем исследовании данные значения можно приравнять к значению `'NaN'`. 

```python
display(data[data['user_score'] == 'tbd']) # срез таблицы по значению 'tbd' в оценке пользователей
```

```python
data['user_score'] = data['user_score'].replace('tbd', float('NaN')) 
# замена значения 'tbd' на значение 'NaN' в оценке пользователей
```

```python
print(data['user_score'].isna().sum()) # подсчёт пропусков после замены значения 'tbd' в столбце 'user_score'
```

После проведения такой замены количество пропусков увеличилось до 9027. 

Замену пропущенный значений в столбце `user_score` проведем по аналогии со столбцом `critic_score` в два этапа: 
1. Создадим новую переменную `transformed_user_score`, в которой сгруппируем данные по названию игр и подсчитаем среднюю оценку пользователей для каждой группы, предварительно её огруглив до ближайшего значения до одного знака после запятой. Произведем замену пропусков значением из этой переменной при совпадении названия игры.
2. Для замены оставшихся 7928 пропусков используем значение -1. При выгрузке уникальных значений была выявлена только одна игра с 0 рейтингом (скорее всего оценка игры отсутсвует или указана неверно). Соответственно, -1 даст возможность учитывать при анализе полный спектр строк, сразу определить строки замены и минимизировать искажение. 

Но для дальнейших преобразований первоначально потребуется замена типа данных на вещественное число с текстового значения.

```python
data['user_score'] = data['user_score'].astype('float') 
# перевод значений в столбце 'user_score' из object в тип данных float
```

```python
transformed_user_score = round(data.groupby('name')['user_score'].transform('mean'), 1)
# вычисление средней оценки для названия каждой игры с округлением до ближайшего значения до одного знака после запятой
data['user_score'] = data['user_score'].fillna(transformed_user_score) 
# замена пропусков в столбце 'user_score' на среднюю оценку для названия каждой игры
```

```python
print(data['user_score'].isna().sum()) # подсчёт пропусков после первой замены в столбце 'critic_score'
```

```python
display(data['user_score'].value_counts().sort_values())
# подсчёт уникальных значений оценок пользователей с сортировкой по возрастанию
```

```python
display(data[data['user_score'] == 0]) # срез таблицы по нулевому значения оценки пользователей
```

## Составление портрета пользователя каждого региона (NA, EU, JP)

###  Самые популярные платформы

Для начала для каждого региона (Северной Америки, Европы и Японии) определим топ-5 самых популярных игровых платформ. 

```python
data_columns = ['na_sales', 'eu_sales', 'jp_sales'] #  создание списка региональных продаж
for column in data_columns: # создание цикла
    platform_top = (
        data_profitable_platform.pivot_table(index='platform', values=column, aggfunc='sum')
        .sort_values(by=column, ascending=False).head()
        ) # формирование сводной таблицы platform_top из первых 5 платформ с продажами, отсортированных по убыванию
    platform_top['ratio'] = platform_top[column]/platform_top[column].sum()
    # сохранение в столбец 'ratio' соотношения региональных продаж платформ к общей сумме продаж в регионе 
    platform_top['ratio'].plot(kind='bar', figsize=(4,4), legend=False, title=f'Топ-5 платформ {column}')
    plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
    plt.xlabel('Платформа')
    plt.ylabel('Доля продаж платформ в регионе')
    display(platform_top) # отображение полученной таблицы platform_top
    print()
    plt.show() # отображение столбчатой диаграммы   
```

На графиках и в сводных таблицах хорошо видно, что:

##  Проверка гипотез и пояснения

Проведем проверку двух гипотез:
1. средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
2. средние пользовательские рейтинги жанров Action и Sports разные.

### Проверка гипотезы: средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем нулевую и альтернативную гипотезы:
* H₀ - средние пользовательские рейтинги платформ Xbox One и PC равны;
* H₁ - средние пользовательские рейтинги платформ Xbox One и PC различаются.

```python
data_xone = data_profitable_platform[(data_profitable_platform['platform'] == 'XOne')
                                    & (data_profitable_platform['user_score'] != -1)]['user_score'] 
# выборка средней оценки пользователей платформы Xbox One
display(data_xone) # получение выборки data_xone
```

```python
data_pc = data_profitable_platform[(data_profitable_platform['platform'] == 'PC')
                                    & (data_profitable_platform['user_score'] != -1)]['user_score'] 
# выборка средней оценки пользователей платформы PC
display(data_pc) # получение выборки data_pc
```

```python
alpha = .05 # критический уровень статистической значимости
            # если p-value окажется меньше него - отвегнем гипотезу

results = st.ttest_ind(
    data_xone,
    data_pc,
    equal_var = False) # проверка гипотезы о равенстве средней оценки пользователей платформ Xbox One и PC
                       # выборки разного размера, поэтому equal_var = False

print('p-значение: ', results.pvalue) # вывод на экран получившегося p-значения

if results.pvalue < alpha: # сравнение получившегося p-значения с заданным уровнем статистической значимости
    print("Отвергаем нулевую гипотезу: средние пользовательские рейтинги платформ Xbox One и PC различаются")
else:
    print("Не получилось отвергнуть нулевую гипотезу: средние пользовательские рейтинги платформ Xbox One и PC равны")
print(f'Cредняя оценка пользователей платформы Xbox One: {round(data_xone.mean(),2)}') 
print(f'Cредняя оценка пользователей платформы PC: {round(data_pc.mean(),2)}') 
```

Статистический тест показал, что гипотезу о равенстве среднего пользовательского рейтинга для XOne и PC не получилось отвергнуть. Если бы мы утверждали, что средние рейтинги для этих платформ различаются, мы бы допустили ошибку с вероятностью 5%.

### Проверка гипотезы: средние пользовательские рейтинги жанров Action и Sports разные

Сформулируем нулевую и альтернативную гипотезы:
* H₀ - средние пользовательские рейтинги жанров Action и Sports равны;
* H₁ - средние пользовательские рейтинги жанров Action и Sports разные.

```python
data_action = data_profitable_platform[(data_profitable_platform['genre'] == 'Action')
                                    & (data_profitable_platform['user_score'] != -1)]['user_score'] 
# выборка средней оценки пользователей жанра Action
display(data_action) # получение выборки data_action
```

```python
data_sports = data_profitable_platform[(data_profitable_platform['genre'] == 'Sports')
                                    & (data_profitable_platform['user_score'] != -1)]['user_score'] 
# выборка средней оценки пользователей жанра Sports
display(data_sports) # получение выборки data_sports
```

```python
alpha = .05 # критический уровень статистической значимости
            # если p-value окажется меньше него - отвегнем гипотезу

results = st.ttest_ind(
    data_action,
    data_sports,
    equal_var = False) # проверка гипотезы о равенстве средней оценки пользователей жанров Action и Sports
                       # выборки разного размера, поэтому equal_var = False

print('p-значение: ', results.pvalue) # вывод на экран получившегося p-значения

if results.pvalue < alpha: # сравнение получившегося p-значения с заданным уровнем статистической значимости
    print("Отвергаем нулевую гипотезу: средние пользовательские рейтинги жанров Action и Sports различаются")
else:
    print("Не получилось отвергнуть нулевую гипотезу: средние пользовательские рейтинги жанров Action и Sports равны")
print(f'Cредняя оценка пользователей жанра Action: {round(data_action.mean(),2)}') 
print(f'Cредняя оценка пользователей жанра Sports: {round(data_sports.mean(),2)}') 
```

Исходная гипотеза не подтвердилась: можно утверждать с уверенностью почти 100%, что средние пользовательские рейтинги для жанров Action и Sports различаются. При выборе между продвижением игр двух жанров, при прочих равных условиях лучше отдать предпочтение жанру Action.
